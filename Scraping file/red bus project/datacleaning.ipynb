```python
data['Rating'] = data['Rating'].astype(str)
data['Rating'] = data['Rating'].str.extract(r'(\d+\.\d+)').astype("float64")
data['Rating'] = data['Rating'].fillna(0.0)
data["Rating"] = pd.to_numeric(data["Rating"], errors="coerce")
```

```python
data['Seatavailable'] = data['Seatavailable'].astype(str)
data['Seatavailable'] = data['Seatavailable'].str.extract(r'(\d+)').astype(float).astype('Int64')

data["Seatavailable"]=pd.to_numeric(data["Seatavailable"],errors="coerce")
```

```python
data['Price'] = data['Price'].astype(str)
data['Price'] = data['Price'].str.replace('INR ', '')
data['Price'] = data['Price'].astype("float64")
```

```python
data['Duration'] = data['Duration'].astype(str)

# Apply a lambda function directly
data['Duration'] = data['Duration'].apply(
    lambda x: " ".join(re.search(r"(\d+)[hH]\s?(\d+)[mM]", x).groups()) if re.search(r"(\d+)[hH]\s?(\d+)[mM]", x) else None
)
```

```python
data.dropna(subset=["Rating"], inplace=True)
data.dropna(subset=["Bustype"], inplace=True)
data.dropna(subset=["Seatavailable"], inplace=True)
data.dropna(subset=["Price"], inplace=True)
data.dropna(subset=["Duration"], inplace=True)
```

```python
data['Busname'] = data['Busname'].astype(str)
data['Bustype'] = data['Bustype'].astype(str)
data['Duration'] = data['Duration'].astype(str)
data['Route_name'] = data['Route_name'].astype(str)
data['Route_link'] = data['Route_link'].astype(str)
```

```python


# Establish a connection to the MySQL database
conn = mysql.connector.connect(
    host="localhost",      
    user="root",  
    password="1234",    
    database="red_bus_details" 
)

# Create a cursor to execute SQL queries
cursor = conn.cursor(buffered=True)
```

```python

create_table_query = """
CREATE TABLE IF NOT EXISTS RED_BUS_DETAILS2 (
    id INT AUTO_INCREMENT PRIMARY KEY,
    Busname VARCHAR(255),
    Bustype VARCHAR(255),
    Starttime TIME,
    Endtime TIME,
    Duration VARCHAR(255),
    Price DECIMAL(10,2),
    Rating FLOAT(3,1),
    Seatavailable BIGINT,
    Route_name VARCHAR(255),
    Route_link VARCHAR(255)
);
"""
try:
    cursor.execute(create_table_query)
    print("Table 'RED_BUS1' created successfully!")
except mysql.connector.Error as err:
    print(f"Error: {err}")


```

```python


cursor = conn.cursor()

# Define the SQL query for inserting data
sql = """
INSERT INTO RED_BUS_DETAILS2 (
    Busname, Bustype, Starttime, Endtime, Duration, 
    Price, Rating, Seatavailable, Route_name, Route_link
) 
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Insert data row by row
for _, row in data.iterrows():
    values = (
        row['Busname'], row['Bustype'], row['Starttime'], row['Endtime'], 
        row['Duration'], row['Price'], row['Rating'], 
        row['Seatavailable'], row['Route_name'], row['Route_link']
    )
    cursor.execute(sql, values)

# Commit changes and close the connection
conn.commit()


print("Data inserted into the `RED_BUS_DETAILS2` table successfully!")
```
