```python
# import libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv(r"df_K.csv")
df1
```

```python

# Initialize WebDriver
driver = webdriver.Chrome()
driver.maximize_window()

# Initialize lists to store scraped data
Bus_name_K= []
Bus_type_K = []
Start_time_K= []
End_time_K= []
Rating_K= []
duration_K= []
price_K=[]
Seat_available_K = []
Route_name_K=[]
Route_link_K= []

for i, j in df1.iterrows():
    route_name = j["route_names"]
    route_link = j["route_links"]
    
    # Navigate to the link
    driver.get(route_link)
    time.sleep(5)  # You can adjust this based on your page load time

    # Function to scrape bus details
    def scrape_buses():
        scrolling = True
        retries = 0

        while scrolling and retries < 5:
            old_page_source = driver.page_source
            ActionChains(driver).send_keys(Keys.PAGE_DOWN).perform()
            time.sleep(3)

            new_page_source = driver.page_source
            if old_page_source == new_page_source:
                retries += 1
            else:
                retries = 0

            if retries >= 5:
                scrolling = False

    
        try:
            # Scrape details after interacting with 'View Bus' buttons
            bus_names = driver.find_elements(By.CSS_SELECTOR, 'div[class="travels lh-24 f-bold d-color"]')
            bus_types = driver.find_elements(By.CSS_SELECTOR, 'div[class="bus-type f-12 m-top-16 l-color evBus"]')
            durations = driver.find_elements(By.CSS_SELECTOR, 'div[class="dur l-color lh-24"]')
            start_times = driver.find_elements(By.CSS_SELECTOR, 'div[class="dp-time f-19 d-color f-bold"]')
            end_times = driver.find_elements(By.CSS_SELECTOR, 'div[class="bp-time f-19 d-color disp-Inline"]')
            prices = driver.find_elements(By.XPATH, '//div[@class="fare d-block"]')
            seat_availabilities = driver.find_elements(By.XPATH, '//div[contains(@class,"seat-left")]')
            star_ratings=  driver.find_elements(By.CSS_SELECTOR,'div[class="column-six p-right-10 w-10 fl"]')
            
            # Append scraped data to lists
            Bus_name_K.extend([i.text.strip() for i in bus_names])
            Bus_type_K.extend([i.text.strip() for i in bus_types])
            Start_time_K.extend([i.text.strip() for i in start_times])
            End_time_K.extend([i.text.strip() for i in end_times])
            Rating_K.extend([i.text.strip() for i in star_ratings])
            duration_K.extend([i.text.strip() for i in durations])
            price_K.extend([i.text.strip() for i in prices])
            Seat_available_K.extend([i.text.strip() for i in seat_availabilities])
            Route_name_K.extend([route_name] * len(bus_names))  # Append route_name for each bus
            Route_link_K.extend([route_link] * len(bus_names))  # Append route_link for each bus

        except Exception as e:
            print(f"Error scraping bus details: {e}")

    # Execute the scraping function
    scrape_buses()

# Close the WebDriver
driver.quit()
```
